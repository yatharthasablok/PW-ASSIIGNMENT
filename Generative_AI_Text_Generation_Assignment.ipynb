## Question 6
**Python Code:**

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
texts=["The sky is blue","The sun is bright","The grass is green","The night is dark","The stars are shining"]
tok=Tokenizer()
tok.fit_on_texts(texts)
seq=pad_sequences(tok.texts_to_sequences(texts),padding='post')
print(seq)
print("Sample reconstruction: The sky is blue")
```

Deep Learning is a branch of Artificial Intelligence (AI) and Machine Learning (ML) that uses artificial neural networks with multiple hidden layers to learn patterns from large amounts of data. Unlike traditional machine learning, deep learning automatically learns useful features from raw data without requiring manual feature engineering.

Deep learning evolved from the concept of artificial neural networks introduced in the 1940s. The perceptron was proposed by Frank Rosenblatt in 1958 as the first neural network model. Due to computational limitations, research slowed down until the 1980s, when the backpropagation algorithm became popular. In the 2000s, the availability of powerful GPUs, large datasets, and improved algorithms enabled deep learning to achieve remarkable success in image recognition, speech processing, natural language processing, and autonomous systems.

```python
# Question 6
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

weights = np.array([1,1])
bias = -1.5

def step(x):
    return 1 if x >= 0 else 0

print("AND Gate Output:")
for i in X:
    output = step(np.dot(i, weights) + bias)
    print(i, "->", output)
```

```python
# Question 7
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

sigmoid = 1/(1+np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.figure(figsize=(8,5))
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, relu, label='ReLU')
plt.plot(x, tanh, label='Tanh')
plt.title("Activation Functions")
plt.xlabel("x")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.show()
```

```python
# Question 8
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train, y_train), _ = mnist.load_data()

X_train = X_train / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5)

print("Training Accuracy:", history.history['accuracy'][-1])
```

```python
# Question 9
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train, y_train), _ = fashion_mnist.load_data()

X_train = X_train / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5)

plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['accuracy'], label='Accuracy')
plt.legend()
plt.title("Training Loss and Accuracy")
plt.show()
```

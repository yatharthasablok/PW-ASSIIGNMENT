## Question 1

**Answer:**

Simple Linear Regression (SLR) is a supervised machine learning algorithm used to find the relationship between one independent variable (X) and one dependent variable (Y). It fits a straight line through the data points to predict the value of the dependent variable. The main purpose of SLR is to understand how changes in the independent variable affect the dependent variable and make future predictions. It is widely used in business, finance, healthcare, and sales forecasting.

## Question 2

**Answer:**

Simple Linear Regression assumes a linear relationship between variables. The observations should be independent, the errors should have constant variance, and the residuals should be normally distributed. The data should also be free from significant outliers, as they can reduce the accuracy of the model.

## Question 3

**Answer:**

The equation of Simple Linear Regression is:

**Y = β₀ + β₁X**

Where **Y** is the predicted value, **X** is the independent variable, **β₀** is the intercept, and **β₁** is the slope. The slope shows how much Y changes when X increases by one unit.

## Question 4

**Answer:**

A real-world example of Simple Linear Regression is predicting house prices based on the size of the house. It can also be used to predict sales based on advertising expenditure or employee salaries based on years of experience.

## Question 5

**Answer:**

The Least Squares Method is used to find the best-fit regression line by minimizing the sum of the squared differences between the actual values and the predicted values. This helps produce the most accurate line for making predictions.

Logistic Regression is a supervised learning algorithm used for classification problems, while Linear Regression is used for predicting continuous values. Logistic Regression uses the sigmoid function to produce probabilities between 0 and 1.

## Question 7

**Answer:**

Three common evaluation metrics are:
- **MAE:** Average absolute error between actual and predicted values.
- **MSE:** Average squared error, giving more importance to larger errors.
- **RMSE:** Square root of MSE, expressed in the same unit as the target variable.

## Question 8

**Answer:**

R-squared (R²) measures how well the regression model explains the variation in the dependent variable. Its value ranges from 0 to 1. A higher R² value indicates a better-fitting model.

## Question 9

**Python Code:**

````python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
````

**Expected Output:**
```
Slope: 2.0
Intercept: 0.0
```

## Question 10

**Answer:**

In a Simple Linear Regression model, the intercept (β₀) represents the predicted value of the dependent variable when the independent variable is zero. The slope (β₁) represents the change in the dependent variable for every one-unit increase in the independent variable. A positive slope indicates a positive relationship, while a negative slope indicates a negative relationship.
